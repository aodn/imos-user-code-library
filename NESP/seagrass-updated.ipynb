## Connecting to NESP 5.9 Datasets

NESP 5.9 datasets can be connected to via a pyarrow `S3FileSystem`.

The `S3FileSystem` manages the connection to the NESP 5.9 datasets stored in S3, resulting in a `pyarrow.Dataset` handle that we may use for efficient loading and subsetting of the dataset.

Note that a dataset can be comprised of a single parquet file or partitioned (into separate parquet fragments/files).

This step also makes the dataset schema available and file structure available allowing for pre-filtering without downloading the entire dataset.

Finally, we may estimate the in memory dataset size by sampling the dataset. If the dataset is small, we may use a `DataFrame` straight away. Otherwise if the dataset is too large, we must pre-aggregate and or filter using a `LazyFrame` before loading the dataset into memory.

````python
import pyarrow
import pyarrow.fs
import pyarrow.dataset
import util

# Construct the anonymous file system responsible for reading from the public S3 bucket
FILE_SYSTEM = pyarrow.fs.S3FileSystem(
    region="ap-southeast-2", 
    anonymous=True,
)

# Create the dataset connection
# By convention, datasets are labelled `ds`
ds = pyarrow.dataset.dataset(
    source="data-uplift-public/stored/datauplift/seagrass/seagrass.parquet",
    filesystem=FILE_SYSTEM,
)

# The comprising files of the dataset can be inspected
print(list(ds.get_fragments()))

# The schema of the dataset can be inspected
util.print_schema_rich_table(
    schema=ds.schema,
    metadata_keys=["definition", "units", "long_name", "resolution"]
)

# For the full table scheme use native pyarrow representation
# rich.print(ds.schema)

# Estimate dataset size
estimated_megabytes = util.estimate_dataset_size(
    ds=ds,
    n_samples=10_000,
)
print(f"DS estimated size: {round(estimated_megabytes, 2)}MB")
````

## Loading Data
The dataset handle is our connection to data in S3. We can use this handle to load data directly into a `DataFrame` or `LazyFrame`

### DataFrame
For smaller datasets the `DataFrame` is preferrable. This loads the entire dataset into memory ready for analysis.

### LazyFrame
For larger datasets the `DataFrame` may not fit into memory. In these cases the `LazyFrame` is ideal as it allows the dataset to be filtered and or aggregated prior to loading the dataset, allowing it to fit into memory, ready for analysis.

````python
import polars
import polars_h3

df = polars.DataFrame(
    data=ds.to_table(),
)
````

## Observation Effort, Organism Quantity and Richness

### Observation Effort and Organism Quantity
**Observation Effort** is the number of occurrence records while **Organism Quantity** is the recorded count of a species occurrence.

**Observation Effort** must be taken into account when analysing trends; fluctuations in observation effort tend to distort the reported occurrences.

### Richness
Species **Richness** refers to the count of different species present in a particular area. While Organism Quantity tells us "how many" individuals there are, **Richness** tells us "how many types" of organisms coexist.

Typically the **Richness** of an area correlates to the number of studies but eventually plateus and this plateu is an indication that the studied region is well understood.

````python
import altair

x = altair.X(
    "year:Q", 
    title="Year",
)

yearly_stat_df = (
    df
    .group_by(
        polars.col("startYear").alias("year"),
    )
    .agg(
        polars.col("datasetName").unique().alias("datasets"),
        polars.len().alias("nRecords"),
        polars.col("occurrenceStatus").replace({
            "present": "1",
            "absent": "0",
        }).cast(polars.Int64).sum().alias("organismQuantity")
    )
    .with_columns(
        polars.col("datasets").list.join(", "),
        polars.col("datasets").list.len().alias("nDatasets"),
    )
    .sort(
        polars.col("year"),
    )
)
dataset = (
    altair.Chart(
        data=yearly_stat_df,
        title=altair.Title(
            "Dataset Quantity Over Time",
            subtitle="The number of datasets fluctuates over the aggregated dataset",
            anchor="start",
            orient="top",
            offset=20,
            frame="group",
        ),
        width=1000,
        height=200
    )
    .mark_bar(
        point=True,
        interpolate="monotone",
    )
    .encode(
        x=x,
        y=altair.Y("nDatasets:Q"),
        tooltip=["year", "datasets", "nRecords"],
    )
)
observation = (
    altair.Chart(
        data=yearly_stat_df,
        title=altair.Title(
            "Observations Over Time",
            subtitle="Observation effort fluctuates over the aggregated dataset and is correlated to the number of underlying datasets",
            anchor="start",
            orient="top",
            offset=20,
            frame="group",
        ),
        width=1000,
        height=200
    )
    .mark_bar(
        point=True,
        interpolate="monotone",
    )
    .encode(
        x=x,
        y="nRecords:Q",
        tooltip=["year", "datasets", "nRecords"],
    )
)
quantity = (
    altair.Chart(
        data=yearly_stat_df,
        title=altair.Title(
            "Organism Quantity Over Time",
            subtitle="Organism quantity similarly fluctuates over the aggregated dataset and is correlated to the number of underlying datasets",
            anchor="start",
            orient="top",
            offset=20,
            frame="group",
        ),
        width=1000,
        height=200
    )
    .mark_bar(
        point=True,
        interpolate="monotone",
    )
    .encode(
        x=x,
        y=altair.Y("organismQuantity:Q"),
        tooltip=["year", "datasets", "nRecords"],
    )
)
richness_df = (
    df
    .group_by(
        polars.col("scientificNameID")
    )
    .agg(
        polars.struct(
            polars.col("startYear").min().alias("year"),
            polars.col("datasetName").alias("dataset"),
        ).first().alias("year_dataset"),
    )
    .sort(
        polars.col("year_dataset"),
    )
    .group_by(
        polars.col("year_dataset").struct.field("year")
    )
    .agg(
        polars.col("year_dataset").struct.field("dataset").unique().alias("datasets"),
        polars.col("scientificNameID").count().alias("nSpecies"),
        polars.col("scientificNameID").unique().cast(polars.String),
    )
    .sort("year")
    .with_columns(
        polars.col("scientificNameID").list.join(", "),
        polars.col("datasets").list.join(", "),
        polars.col("nSpecies").cum_sum().alias("richness"),
    )
)
richness = (
    altair.Chart(
        data=richness_df,
        title=altair.Title(
            "Species Richness Over Time",
            subtitle="Species richness has not yet plateued",
            anchor="start",
            orient="top",
            offset=20,
            frame="group",
        ),
        width=1000,
        height=500
    )
    .mark_area(
        point=True,
        interpolate="monotone",
    )
    .encode(
        x=x,
        y=altair.Y("richness").title("Richness"),
        tooltip=["year", "richness", "datasets", "nSpecies", "scientificNameID"],
    )
)

altair.vconcat(
    dataset,
    observation,
    quantity,
    richness,
)
````

````python
mesoscale_water_type_df = (
    df
    .with_columns(
        polars.col("australianMarineRegionsTags")
        .str.split("|")
        .list.filter(
            polars.element().str.contains("IMCRA:Mesoscale:Water Type")
        )
    )
    .filter(
        polars.col("australianMarineRegionsTags").is_not_null(),
    )
    .filter(
        polars.col("australianMarineRegionsTags").list.len().gt(0),
    )
    .explode(
        polars.col("australianMarineRegionsTags"),
    )
    .with_columns(
        polars.col("australianMarineRegionsTags")
        .str.strip_prefix("IMCRA:Mesoscale:Water Type:")
        .alias("mesoscaleWaterType"),
    )
    .sort(
        polars.col("startYear"),
        polars.col("startMonth"),
        polars.col("startDay")
    )
)
mesoscale_water_type_yearly_stat_df = (
    mesoscale_water_type_df
    .group_by(
        polars.col("startYear").alias("year"),
        polars.col("mesoscaleWaterType"),
    )
    .agg(
        polars.col("datasetName").unique().alias("datasets"),
        polars.len().alias("nRecords"),
        polars.col("occurrenceStatus").replace({
            "present": "1",
            "absent": "0",
        }).cast(polars.Int64).sum().alias("organismQuantity")
    )
    .sort(
        polars.col("year"),
        polars.col("mesoscaleWaterType"),
    )
    .with_columns(
        polars.col("datasets").list.join(", "),
        polars.col("datasets").list.len().alias("nDatasets"),
    )
)
mesoscale_water_type_richness_df = (
    mesoscale_water_type_df
    .group_by(
        polars.col("scientificNameID"),
        polars.col("mesoscaleWaterType"),
    )
    .agg(
        polars.struct(
            polars.col("startYear").min().alias("year"),
            polars.col("datasetName").alias("dataset"),
        ).first().alias("year_dataset"),
    )
    .sort(
        polars.col("year_dataset"),
    )
    .group_by(
        polars.col("year_dataset").struct.field("year"),
        polars.col("mesoscaleWaterType"),
    )
    .agg(
        polars.col("year_dataset").struct.field("dataset").unique().alias("datasets"),
        polars.col("scientificNameID").count().alias("nSpecies"),
        polars.col("scientificNameID").unique().cast(polars.String),
    )
    .sort("year")
    .with_columns(
        polars.col("scientificNameID").list.join(", "),
        polars.col("datasets").list.join(", "),
        polars.col("nSpecies").cum_sum().over("mesoscaleWaterType").alias("richness"),
    )
)
domain = (
    mesoscale_water_type_yearly_stat_df["year"].min(),
    mesoscale_water_type_yearly_stat_df["year"].max(),
)
x = altair.X(
    "year:Q", 
    title="Year",
).scale(domain=domain)

for mesoscaleWaterType in mesoscale_water_type_yearly_stat_df["mesoscaleWaterType"].unique().sort():

    dataset = (
        altair.Chart(
            data=mesoscale_water_type_yearly_stat_df.filter(
                polars.col("mesoscaleWaterType").eq(mesoscaleWaterType),
            ),
            title=altair.Title(
                "Dataset Quantity Over Time",
                anchor="start",
                orient="top",
                offset=20,
                frame="group",
            ),
            width=1000,
            height=200
        )
        .mark_bar(
            point=True,
            interpolate="monotone",
        )
        .encode(
            x=x,
            y=altair.Y("nDatasets:Q"),
            tooltip=["year", "datasets", "nRecords"],
        )
    )

    observation = (
        altair.Chart(
            data=mesoscale_water_type_yearly_stat_df.filter(
                polars.col("mesoscaleWaterType").eq(mesoscaleWaterType),
            ),
            title=altair.Title(
                "Observations Over Time",
                anchor="start",
                orient="top",
                offset=20,
                frame="group",
            ),
            width=1000,
            height=200
        )
        .mark_bar(
            point=True,
            interpolate="monotone",
        )
        .encode(
            x=x,
            y="nRecords:Q",
            tooltip=["year", "datasets", "nRecords"],
        )
    )

    quantity = (
        altair.Chart(
            data=mesoscale_water_type_yearly_stat_df.filter(
                polars.col("mesoscaleWaterType").eq(mesoscaleWaterType),
            ),
            title=altair.Title(
                "Organism Quantity Over Time",
                anchor="start",
                orient="top",
                offset=20,
                frame="group",
            ),
            width=1000,
            height=200
        )
        .mark_bar(
            point=True,
            interpolate="monotone",
        )
        .encode(
            x=x,
            y=altair.Y("organismQuantity:Q"),
            tooltip=["year", "datasets", "nRecords"],
        )
    )

    richness = (
        altair.Chart(
            data=mesoscale_water_type_richness_df.filter(
                polars.col("mesoscaleWaterType").eq(mesoscaleWaterType),
            ),
            title=altair.Title(
                "Species Richness Over Time",
                anchor="start",
                orient="top",
                offset=20,
                frame="group",
            ),
            width=1000,
            height=500
        )
        .mark_area(
            point=True,
            interpolate="monotone",
        )
        .encode(
            x=altair.X("year").title("Year"),
            y=altair.Y("richness").title("Richness"),
            tooltip=["year", "richness", "datasets", "nSpecies", "scientificNameID"],
        )
    )

    altair.vconcat(
        dataset,
        observation,
        quantity,
        richness,
    ).properties(
        title=altair.Title(
            text=mesoscaleWaterType,
            fontSize=20,
            anchor="start",
            offset=20,
            frame="group",
            
        ),
    ).show()
````

## H3 Aggregation and Mapping
The NESP 5.9 datasets are aggregated data products that cover large regions and time periods.

It is useful to first aggregate the datasets so that we may understand their temporal-spatial bounds.

### H3 Indexing
The `h3` library allows us to spatially index datasets at different [resolutions](https://h3geo.org/docs/core-library/restable#average-area-in-km2) (zoom levels) according to the record latitude and longitude.

Additionally, the `polars_h3` library offers a polars style interface for faster tagging with `polars.DataFrames`:
```python

# Add a h3 index to any dataframe with latitude and longitude columns
df = df.with_columns(

    # Polars Expression Shorthand to access the `polars_h3` namespace
    polars_h3.latlng_to_cell(

        # Specify the latitude and longitude columns
        lat=polars.col("decimalLatitude"),
        lng=polars.col("decimalLongitude"),

        # Specify the resolution
        resolution=5,

        # Specify the index return type, can be `polars.Int64` or `polars.String`
        return_dtype=polars.String,
    )
    
    # Alias to `h3Index`
    .alias("h3Index"),
)
```

### H3 Mapping
The `pydeck` library offers a built in [`H3HexagonLayer`](https://deck.gl/docs/api-reference/geo-layers/h3-hexagon-layer) layer type that greatly reduces complexity in mapping H3 Shapes:

```python

# Generate a h3 hexagon layer displayable on a pydeck map
pydeck.Layer(

    # Specify this is a H3 Hexagon Layer
    "H3HexagonLayer",

    # Unfortunately no polars dataframe support, so cast to `pandas.DataFrame`
    df.to_pandas(use_pyarrow_extension_array=True),

    # Specify the hexagon
    get_hexagon="h3Index",

)
```
This is rendered with a [`pydeck.Deck`](https://deck.gl/docs/api-reference/core/deck):
```python

# Generate a deck with h3 layers
deck = pydeck.Deck(

    # Supply the h3 layers
    layers=h3_layers,
)
deck.show()
```

````python
import pydeck

# Aggregate and generate h3 layers
h3_layers = util.generate_pydeck_hexagon_layers(
    df=(

    # H3 Aggregated dataframe with dataset and number of records attributes
    df

    .group_by(
        polars.col("h3Index"),
    )
    .agg(
        polars.len().alias("n_records"),
        polars.col("datasetName"),
    )
    .select(
        polars.col("h3Index"),
        polars.col("n_records"),
        polars.col("datasetName").list.unique().list.sort().list.join(", ").alias("datasets"),
    )
),
    aggregate_column_name="n_records",
    n_quantiles = 10,
    color_palette="plasma",
)

# Run the map
deck = pydeck.Deck(
    layers=h3_layers,
    map_style=pydeck.map_styles.LIGHT_NO_LABELS,
    tooltip=util.TOOLTIP,
    initial_view_state=util.generate_view_state(df),
)
deck.show()
````

### Exploring Available Marine Regions Tags

Extracting the available Australian Marine Regions Tags requires spliting the delimited column:

````python
# Extract the unique tags
australian_marine_regions_tags = df["australianMarineRegionsTags"].str.split("|").explode().value_counts(sort=True).drop_nulls()

# Maniuplate polars display config to display full tag names
with polars.Config(
    fmt_str_lengths=128,
    tbl_rows=20
):
    display(australian_marine_regions_tags)
````

````python
# Filter to the commonwealth marine regions
australian_marine_regions_df = df.filter(
    polars.col("australianMarineRegionsTags").is_null().not_()
)

# Aggregate and generate h3 layers
h3_layers = util.generate_pydeck_hexagon_layers(
    df=(
        australian_marine_regions_df

        .group_by(
            polars.col("h3Index"),
        )
        .agg(
            polars.len().alias("n_records"),
            polars.col("datasetName"),
            polars.col("australianMarineRegionsTags"),
        )
        .with_columns(
            polars.col("datasetName").list.unique().list.sort().list.join(", ").alias("datasets"),
            # De-duplicate marine tags
            (
                polars.col("australianMarineRegionsTags")
                # Flatten "|" separated list of strings into single wide "|" separated str
                .list.unique().list.join("|")
                # Split and filter to sorted uniques of wide "|" separated str
                .str.split("|").list.unique().list.sort().list.join("|").alias("tags")
            ),
        )
        .select(
            polars.col("h3Index"),
            polars.col("n_records"),
            polars.col("datasets"),
            polars.col("tags"),
        )
    ),
    aggregate_column_name="n_records",
    n_quantiles = 10,
    color_palette="plasma",
)

# Run the map
deck = pydeck.Deck(
    layers=h3_layers,
    map_style=pydeck.map_styles.LIGHT_NO_LABELS,
    tooltip={
        "html": """
            <div style="font-family: 'Helvetica Neue', Arial, sans-serif; padding: 10px;">
                <b style="font-size: 1.2em;">H3 Index:</b> <code>{h3Index}</code><br/>
                <hr style="margin: 5px 0; border: 0; border-top: 1px solid #ccc;">
                <b>Records:</b> {n_records}<br/>
                <b>Datasets:</b> {datasets}<br/>
                <b>Tags:</b> {tags}<br/>
            </div>
        """,
        "style": {
            "width": "33%",
            "backgroundColor": "#2b2b2b",
            "color": "white",
            "borderRadius": "4px",
            "zIndex": "1000"
        }
    },
    initial_view_state=util.generate_view_state(australian_marine_regions_df)
)
deck.show()
````

### Filtering to a Particular Marine Region
We may filter to a particular requires a using `str.contains` expression:

````python
# Filter to the commonwealth marine regions
tropical_waters_df = df.filter(
    polars.col("australianMarineRegionsTags").str.contains(
        pattern="IMCRA:Provincial:Water Type:Tropical Waters",
        literal=True,
    )
)

# Aggregate and generate h3 layers
h3_layers = util.generate_pydeck_hexagon_layers(
    df=(
        tropical_waters_df
        .with_columns(
            polars_h3.latlng_to_cell(
                lat=polars.col("decimalLatitude"),
                lng=polars.col("decimalLongitude"),
                resolution=7,
                return_dtype=polars.String,
            ).alias("h3Index"),
        )
        .group_by(
            polars.col("h3Index"),
        )
        .agg(
            polars.len().alias("n_records"),
            polars.col("datasetName"),
            polars.col("australianMarineRegionsTags"),
        )
        .with_columns(
            polars.col("datasetName").list.unique().list.sort().list.join(", ").alias("datasets"),
            # De-duplicate marine tags
            (
                polars.col("australianMarineRegionsTags")
                # Flatten "|" separated list of strings into single wide "|" separated str
                .list.unique().list.join("|")
                # Split and filter to sorted uniques of wide "|" separated str
                .str.split("|").list.unique().list.sort().list.join("|").alias("tags")
            ),
        )
        .select(
            polars.col("h3Index"),
            polars.col("n_records"),
            polars.col("datasets"),
            polars.col("tags"),
        )
    ),
    aggregate_column_name="n_records",
    n_quantiles = 10,
    color_palette="plasma",
)

# Run the map
deck = pydeck.Deck(
    layers=h3_layers,
    map_style=pydeck.map_styles.LIGHT_NO_LABELS,
    tooltip=util.TOOLTIP,
    initial_view_state=util.generate_view_state(tropical_waters_df),
    height=1000,
)
deck.show()
````
